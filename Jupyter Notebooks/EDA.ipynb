```python
"""
This notebook has been produced by me, Booy Faassen, but has already been submitted and graded in a previous course
(the research project). However, its visualizations are re-used in this examination.
Author: Booy Faassen
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

label_doc = pd.read_csv("./data/labels.csv", header = None, delim_whitespace=True, names = ["image", "value", "type"])

#print(labels)
#print(labels.shape)
```

```python
""" Count all instances of the different labels and initialize variables """

count_1_mono = 0
count_1_poly = 0

count_0_mono = 0
count_0_poly = 0

count_03_mono = 0
count_03_poly = 0

count_06_mono = 0
count_06_poly = 0

for row in range(label_doc.shape[0]):
    if label_doc.iloc[row]["value"] == 1.0:
        if label_doc.iloc[row]["type"] == "mono":
            count_1_mono += 1
        else:
            count_1_poly += 1
    elif label_doc.iloc[row]["value"] == 0.0:
        if label_doc.iloc[row]["type"] == "mono":
            count_0_mono += 1
        else:
            count_0_poly += 1
    elif label_doc.iloc[row]["value"] == 0.3333333333333333:
        if label_doc.iloc[row]["type"] == "mono":
            count_03_mono += 1
        else:
            count_03_poly += 1
    else:
        if label_doc.iloc[row]["type"] == "mono":
            count_06_mono += 1
        else:
            count_06_poly += 1


labels = ["1.0", "0.6", "0.3", "0.0"]
mono_counts = [count_1_mono, count_06_mono, count_03_mono, count_0_mono]
poly_counts = [count_1_poly, count_06_poly, count_03_poly, count_0_poly]

mono_count = 0
for count in mono_counts:
    mono_count += count

poly_count = 0
for count in poly_counts:
    poly_count += count
```

```python
fig = plt.figure()

rows = 1
columns = 2

fig, ax = plt.subplots(rows, columns)

width = 0.35
x = np.arange(len(labels))

rects1 = ax[1].bar(x - width/2, mono_counts, width, label='Monocrystalline')
rects2 = ax[1].bar(x + width/2, poly_counts, width, label='Polycrystalline')

ax[1].set_ylabel('Count')
ax[1].set_xlabel('Rating')
ax[1].set_title('(b) - Counts of Ratings by type')
ax[1].set_xticks(x, labels)
ax[1].set(ylim=(0, 1100))
ax[1].legend(loc='upper left')

ax[1].bar_label(rects1, padding=1)
ax[1].bar_label(rects2, padding=1)

counts = [count_1_mono + count_1_poly, count_06_mono + count_06_poly, count_03_mono + count_03_poly, count_0_mono + count_0_poly]

rects3 = ax[0].bar(x, counts, width, color='green')

ax[0].set_ylabel('Count')
ax[0].set_xlabel('Rating')
ax[0].set_title('(a) - Counts of Ratings')
ax[0].set_xticks(x, labels)
ax[0].set(ylim=(0, 1700))

ax[0].bar_label(rects3, padding=1)

fig.tight_layout()
fig.set_figwidth(12)
plt.show()
```

```python
fig = plt.figure()

rows = 1
columns = 2

fig, ax = plt.subplots(rows, columns)

labels = ['Monocrystalline', 'Polycrystalline']
counts = [mono_count, poly_count]

width = 0.15
x = np.arange(len(labels))

# First graph
rects1 = ax[0].bar(x, counts, width, color='green')

ax[0].set_ylabel('Count')
ax[0].set_xlabel('Type')
ax[0].set_title('(a) - Counts of types')
ax[0].set_xticks(x, labels)
ax[0].set(ylim=(0, 1700))

ax[0].bar_label(rects1, padding=1)


# Second graph
counts_1 = [count_1_mono, count_1_poly]
counts_06 = [count_06_mono, count_06_poly]
counts_03 = [count_03_mono, count_03_poly]
counts_0 = [count_0_mono, count_0_poly]

rects1 = ax[1].bar(x - width*1.5, counts_1, width, label='1.0')
rects2 = ax[1].bar(x - width/2, counts_06, width, label='0.6')
rects3 = ax[1].bar(x + width/2, counts_03, width, label='0.3')
rects4 = ax[1].bar(x + width*1.5, counts_0, width, label='0.0')


ax[1].set_ylabel('Count')
ax[1].set_xlabel('Type')
ax[1].set_title('(b) - Counts of types by rating')
ax[1].set_xticks(x, labels)
ax[1].set(ylim=(0, 1100))
ax[1].legend(loc='upper left')

ax[1].bar_label(rects1, padding=1)
ax[1].bar_label(rects2, padding=1)
ax[1].bar_label(rects3, padding=1)
ax[1].bar_label(rects4, padding=1)


fig.tight_layout()
fig.set_figwidth(12)
plt.show()
```
